```python

data=data.drop("name",axis=1)
data=data.drop("id",axis=1)
data=data.drop("sentry_object",axis=1)
data=data.drop("orbiting_body",axis=1)
```

```python
y=data['hazardous']
x =data.iloc[:, 0:5]
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=.3)
```

```python
data.hazardous.value_counts().plot(kind='bar')
plt.title('Hazardous')
plt.show()
```

```python
#plot multiple distributions in dataset 
fig= ses.pairplot(x)
```

```python
knn =KNeighbor(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
best_score = knn.score(x_test,y_test)
```

```python
for i in range(1,15):
    knn2 =KNeighbor(n_neighbors=i)
    knn2.fit(x_train,y_train)
    temp_score = knn2.score(x_test,y_test)
    if(temp_score>best_score):
        knn=knn2
        best_score=temp_score
        
```
